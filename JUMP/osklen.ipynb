```python
import os
import pandas as pd

# Caminho para a pasta contendo os arquivos CSV
pasta = r"C:\Users\natal\OneDrive\Documentos\Pessoal\jump\osklen passado-20240401T182841Z-001\osklen passado"

# Lista para armazenar os DataFrames de cada arquivo CSV
dataframes = []

# Iterar sobre os arquivos na pasta
for arquivo in os.listdir(pasta):
    if arquivo.endswith('.csv'):
        caminho_arquivo = os.path.join(pasta, arquivo)
        df = pd.read_csv(caminho_arquivo,encoding = 'latin-1',sep=';')
        df['nome_arquivo'] = arquivo
        dataframes.append(df)

# Concatenar todos os DataFrames em um Ãºnico DataFrame
osklen = pd.concat(dataframes, ignore_index=True)

```

```python
previsao['diferenca_dias'] = (previsao['Vencimento'] - previsao['Emissao']).dt.days
previsao['tot_parcelas'] = round(previsao['Valor Original Total']/previsao['Valor Pago'],0)
previsao['parcela'] = previsao['diferenca_dias'].apply(lambda x: 'A' if x < 40 else ('B' if x < 70 else ('C' if x < 100 else ('D' if x < 140 else ('E' if x < 160 else None)))))
```

```python
previsao_ = previsao.query("parcela == 'A' ")

previsao_ = previsao_.groupby(['Lancamento Mov','Vencimento']).agg({'Nome Clifor': 'max',
                                                         'Nf Saida': 'max',
                                                         'Valor Original Total':'max',
                                                         'Valor Comissao Linha':'sum',
                                                        'parcela':'max',
                                                        'Emissao':'max',
                                                        'tot_parcelas':'max'}).reset_index()

previsao_.rename(columns={'Vencimento':'data_pagamento',
                          'Nome Clifor':'cliente',
                          'Nf Saida':'documento',
                          'Valor Original Total':'valor',
                          'Valor Comissao Linha':'comissao',
                          'Emissao':'data_faturamento'},inplace=True)

previsao_['cliente'] = previsao_.cliente.str.lower()

previsao_['tipo'] = 'previsao'
```

```python
previsao_.tot_parcelas.unique()
```

```python
from datetime import datetime, timedelta
# Criar novas linhas
novas_linhas = []
for index, row in previsao_.iterrows():
    if row['tot_parcelas'] > 1 and row['tot_parcelas'] <= 5:
        for i in range(1, int(row['tot_parcelas'])):
            nova_parcela = row.copy()
            nova_parcela['tot_parcelas'] = nova_parcela['tot_parcelas'] - i
            nova_parcela['parcela'] = chr(ord(nova_parcela['parcela']) + i)
            nova_parcela['data_pagamento'] += timedelta(days=30*i)
            novas_linhas.append(nova_parcela)

# Adicionar novas linhas ao DataFrame
previsao_final = pd.concat([previsao_, pd.DataFrame(novas_linhas)], ignore_index=True)

# Reordenar o DataFrame
previsao_final = previsao_final.sort_values(by=['Lancamento Mov', 'parcela']).reset_index(drop=True)
```

```python
previsao_final
```

```python
## COMPARAÃÃO

previsao_final['mes_ano'] = previsao_final['data_pagamento'].dt.strftime('%m-%Y')
real['mes_ano'] = real['data_pagamento'].dt.strftime('%m-%Y')
previsao_agrupado = previsao_final[['mes_ano','comissao']].groupby('mes_ano').sum()
real_agrupado = real[['mes_ano','comissao']].groupby('mes_ano').sum()
pd.merge(previsao_agrupado,real_agrupado,on='mes_ano')
```
