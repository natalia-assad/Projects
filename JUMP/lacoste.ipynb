```python
import pandas as pd

# Caminho para o arquivo de texto
caminho_arquivo = r"C:\Users\natal\Downloads\lacoste.txt"

# Linhas que você quer pular
linhas_para_pular = list(range(0, 7))  # Linhas de 0 a 7 (incluindo a 7)
linhas_para_pular.extend(list(range(22, 34)))  # Linhas de 22 a 34 (incluindo a 22)

# Ler apenas as linhas desejadas do arquivo de texto usando o Pandas
dados = pd.read_csv(caminho_arquivo, delimiter='\t', encoding='latin1', skiprows=linhas_para_pular, header=None)
```

```python
dados = dados.drop(index=range(3))
```

```python
dados
```

```python
dados[1] = dados[1].str.replace('.','').str.replace(',','.').astype(float)
```

```python
import pandas as pd
nova_linha = []
for index, row in dados.iterrows():
    nova_linha.append({
            'cliente': None,
            'tipo': 'faturamento',
            'documento': None,
            'valor': row[1],
            'comissao': row[1] * 0.07,
            'parcela': None,
            'data_faturamento': row[0],
            'data_pagamento':pd.to_datetime('1900-01-01'),
            'marca':'lacoste'
        })
```

```python
df = pd.DataFrame(nova_linha)
```

```python
df
```

```python
# Find the index of the first occurrence of NaN in column 'A'
nan_index = df[df['data_faturamento'].isnull()].index[0]

# Drop rows following the first occurrence of NaN in column 'A'
df = df.iloc[:nan_index]
```

```python
meses = {
    'jan': 'Jan',
    'fev': 'Feb',
    'mar': 'Mar',
    'abr': 'Apr',
    'mai': 'May',
    'jun': 'Jun',
    'jul': 'Jul',
    'ago': 'Aug',
    'set': 'Sep',
    'out': 'Oct',
    'nov': 'Nov',
    'dez': 'Dec'
}
# Transformar a coluna 'data' para o tipo datetime
df['data_faturamento'] = pd.to_datetime(df['data_faturamento'].str[:3].map(meses) + '-' + df['data_faturamento'].str[-2:], format='%b-%y')
```

```python
import datetime
hoje = datetime.date.today()

ultimo_dia_mes_anterior = hoje.replace(day=1) - datetime.timedelta(days=1)
```

```python
ultimo_dia_mes_anterior
```

```python
from datetime import datetime, timedelta
nova_linha = []
for index, row in df.iterrows():
    for parcela in ['A', 'B', 'C']:
        data_pagamento= row['data_faturamento'] + timedelta(days=31 * (ord(parcela) - ord('A') + 1))
        if pd.Timestamp(data_pagamento) < pd.Timestamp(ultimo_dia_mes_anterior):
            nova_linha.append({
                'cliente': None,
                'tipo': 'pagamentos',
                'documento': None,
                'valor': row['valor'],
                'comissao': row['comissao']/ 3,
                'parcela': parcela,
                'data_faturamento': row['data_faturamento'],
                'data_pagamento':data_pagamento,
                'marca': 'lacoste'
            })

pagamentos = pd.DataFrame(nova_linha)
```
